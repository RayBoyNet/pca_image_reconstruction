```python
class PCA:
    
    def __init__(self,n_components=2):
        self.n_components = n_components
        self.mean = None
        self._components = None
        self.img = None
        self.img_recon = None
        self.recon_error = None
    
    def fit(self,img):
        # mean centering
        self.img = img
        self.mean = img.mean(axis=0)
        img_center = img - self.mean
        # find covariance matrix
        cov = np.cov(img,rowvar=False)
        # calculate eigenvectors and eigenvalues
        eigvalues, eigvectors = np.linalg.eig(cov)
        # arrange eigenvectors in terms of descending eigenvalues
        idxs = np.argsort(eigvalues)[::-1]
        eigvectors = eigvectors[:,idxs]
        self._components = eigvectors[:,:self.n_components]
    
    def transform(self,img):
        img_center = img-self.mean
        return np.real(img_center@self._components)

    def fit_transform(self,img):
        self.fit(img)
        return self.transform(img)
    
    def inverse_transform(self,img):
        self.img_recon = np.real(img@self._components.T + self.mean)
        self.recon_error = np.sqrt(np.sum((self.img-self.img_recon)**2))
        return self.img_recon
    
```

```python
pca = PCA(n_components=50)
img_norm = img.mean(axis=2)/256
img_reduced = pca.fit_transform(img_norm)
img_recon = pca.inverse_transform(img_reduced)
plt.imshow(1-img_recon,cmap='Greys')
```
